# Model 1: Exploration of the best performing model from tunning and cross-validation

# Presence of *E. coli* multi drug resistance

```python
## Generating model data by approrpiately selecting outcome variable
y = 'Ecoli_R'
features_and_label = features
features_and_label.append(y)
model_data = data[features_and_label]
X_train, X_test, y_train, y_test = train_test_split(model_data[cow_parameters+ herd_parameters], model_data[y], 
                                                    test_size=0.30, stratify = model_data[y])
## processing categorical variables for training and testing data
from sklearn import preprocessing
for c in categorical:
    le = preprocessing.LabelEncoder()
    le.fit(data[c])
    X_train[c] = le.transform(X_train[c]) 
    X_test[c] = le.transform(X_test[c])
    model_data[c] = le.transform(model_data[c])
```

```python
#dt_bin_clf = DecisionTreeClassifier(criterion='entropy', max_depth=3,
#                                    random_state=0)
#dt_bin_clf.fit(X_train, y_train)

dt_multi_clf = DecisionTreeClassifier(criterion='entropy', max_depth=2,
                                      random_state=0)
dt_multi_clf.fit(X_train, y_train)

#dt_reg = DecisionTreeRegressor(criterion='mse', max_depth=3,
#                               random_state=0)
#dt_reg.fit(X_train, y_train)
```

```python
#rf_bin_clf = RandomForestClassifier(criterion='entropy', max_depth=10,
#                                    n_estimators=100, random_state=0)
#rf_bin_clf.fit(X_train, y_train)

rf_multi_clf = RandomForestClassifier(criterion='entropy', max_depth=10,
                                      n_estimators=100, random_state=0)
rf_multi_clf.fit(X_train, y_train)

#rf_reg = RandomForestRegressor(criterion='mse', max_depth=10,
#                               n_estimators=100, random_state=0)
#rf_reg.fit(X_train, y_train)
```
