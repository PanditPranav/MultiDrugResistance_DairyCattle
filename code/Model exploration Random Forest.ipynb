```python
from rfpimp import *
def get_importance(data, y, analysis_name, clf):
    ## processing categorical variables for training and testing data
    from sklearn import preprocessing
    for c in categorical:
        le = preprocessing.LabelEncoder()
        le.fit(data[c])
        data[c] = le.transform(data[c]) 

    for c in categorical:
        le = preprocessing.LabelEncoder()
        le.fit(data[c])
        data[c] = le.transform(data[c]) 


    #best_xgb = XGBClassifier(learning_rate= 0.1, colsample_bytree= 0.3, max_depth= 4, gamma= 0.4, min_child_weight= 1)
    from sklearn.ensemble import ExtraTreesClassifier
    #rf = RandomForestClassifier(max_features= 'auto', n_estimators= 1000, criterion= 'gini', max_depth= 5)
    #rf = RandomForestClassifier(criterion =  'gini', max_depth= 30, max_features = 'auto', min_samples_leaf =  4,
    #                            min_samples_split= 2, n_estimators= 200)
    #rf = RandomForestClassifier(criterion= 'entropy', max_depth= 5, max_features= 'auto',
    #                            min_samples_leaf= 5, min_samples_split= 5, n_estimators= 100)
    rf = clf

    X_train, X_test, y_train, y_test = train_test_split(data[cow_parameters+ herd_parameters], data[y], test_size=0.33, random_state=42)
    print ('Fitting started')
    #xgb = best_xgb.fit(X = X_train, y = y_train)
    rf.fit(X = data[cow_parameters+ herd_parameters], y = data[y])
    
    fig, (ax1, ax2, ax3)= plt.subplots(1, 3, figsize = (12, 12))
    imp = pd.DataFrame({'importance':rf.feature_importances_, 
                        'std': np.std([tree.feature_importances_ for tree in rf.estimators_],axis=0)},
                       index=data[cow_parameters+ herd_parameters].columns,)
    I = pd.DataFrame(imp.importance)
    I['Importance'] = I.importance
    I.sort_values('Importance', ascending = False, inplace = True) 
    viz = plot_importances(I,width=6,title="Importance: gini",imp_range=(0, 0.10), ax= ax1)
    imp = importances(rf, data[cow_parameters+ herd_parameters], data[y]) # permutation
    #print(imp)
    viz = plot_importances(imp, width=6, title='Importance: permutation method',imp_range=(0, 0.10), ax= ax2)
    imp = dropcol_importances(rf, data[cow_parameters+ herd_parameters], data[y]) # permutation
    viz = plot_importances(imp, width=6,  title='Importance : drop column method',imp_range=(0, 0.10),ax= ax3)
    plt.tight_layout()
    plt.savefig('C:/Users/Falco/Desktop/directory/Cattle_AMR_2/outputs/RFimp_'+analysis_name+'.svg')
    plt.show()
    return I
```

```python
from sklearn import preprocessing
def normalize_imp(df):
    min_max_scaler = preprocessing.MinMaxScaler()
    x = df[['importance']].values.astype(float)
    x_scaled = min_max_scaler.fit_transform(x)
    df_normalized = pd.DataFrame(x_scaled, index=df.index)
    return df_normalized


all_cows_n = normalize_imp(all_cows)
all_cows_n['model'] = 'All bacteria'
All_cows_commensal_R_n = normalize_imp(All_cows_commensal_R)
All_cows_commensal_R_n['model'] = 'Commensals'
All_cows_Entero_R_n = normalize_imp(All_cows_Entero_R)
All_cows_Entero_R_n['model'] = 'Enterococcus sp.'
All_cows_Ecoli_R_n = normalize_imp(All_cows_Ecoli_R)
All_cows_Ecoli_R_n['model'] = 'E. coli'
SalPos_allMDR_n = normalize_imp(SalPos_allMDR)
SalPos_allMDR_n['model'] = 'Salmonella sp.'

IMP = pd.concat([
    all_cows_n.head(10),
    All_cows_commensal_R_n.head(10),
    All_cows_Entero_R_n.head(10),
    All_cows_Ecoli_R_n.head(10),
    SalPos_allMDR_n.head(10)
],
          axis=0)

#IMP.columns = ['AllPhenotypes', 'Commensals', 'Entero', 'Ecoli', 'Sal']
#IMP= IMP.sort_values(['AllPhenotypes', 'Commensals', 'Entero', 'Ecoli', 'Sal'], ascending = False)
#IMP_P = IMP.T.reset_index()
#IMP_P
```

```python
from sklearn import tree
import pydotplus
from IPython.display import Image 
import collections

#rf = RandomForestClassifier(max_features= 'auto', n_estimators= 1000, criterion= 'gini', max_depth= 5)
#rf = rf.fit(X = data_m[cow_parameters+ herd_parameters], y = data_m['R'])
# initializing the tree model and training it
from sklearn import preprocessing
for c in categorical:
    le = preprocessing.LabelEncoder()
    le.fit(data[c])
    data[c] = le.transform(data[c]) 

for c in categorical:
    le = preprocessing.LabelEncoder()
    le.fit(data[c])
    data[c] = le.transform(data[c])

tree_model = tree.DecisionTreeClassifier(criterion='gini',
                                         splitter = 'random',
                                         max_depth= 10,
                                         min_samples_split = 3,
                                         max_features='sqrt',
                                         min_samples_leaf = 7,
                                         random_state=43)

tree_model = tree_model.fit(X = data[cow_parameters+ herd_parameters], y = data['R_factor'])

# generating a graphic for the tree
dot_data = tree.export_graphviz(tree_model, out_file=None, 
                         feature_names=cow_parameters+ herd_parameters,
                         class_names = ['No resistance', 'AMR', 'MDR'],
                         filled=True, rounded=False,  
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)

colors = ('#66c2a5','#8da0cb','#fc8d62')
#edges = collections.defaultdict(list)

#for edge in graph.get_edge_list():
#    edges[edge.get_source()].append(int(edge.get_destination()))

#for edge in edges:
#    edges[edge].sort()    
#    for i in range(2):
#        dest = graph.get_node(str(edges[edge][i]))[0]
#        dest.set_fillcolor(colors[i])
nodes = graph.get_node_list()
        
for node in nodes:
    if node.get_name() not in ('node', 'edge'):
        values = tree_model.tree_.value[int(node.get_name())][0]
        #color only nodes where only one class is present
        if max(values) == sum(values):    
            node.set_fillcolor(colors[np.argmax(values)])
        #mixed nodes get the default color
        else:
            node.set_fillcolor('#ffffff')
#plt.show()            
graph.write_svg("C:/Users/Falco/Desktop/directory/Cattle_AMR_2/outputs/dtree_all_cows.svg")
```
